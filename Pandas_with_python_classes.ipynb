# Creating class for performing pandas method on file

- Basic functionality on dataframe

1. dtype
2. ndim
3. size
4. head
5. tail
6. info
7. axes
8. empty
9. shape
10. sample
11. T (Transpose)

```python
#importing module
import pandas as pd

#File path for functionalits of panda
file = "F:\\PGD_Data_Science\\Lecture_video\\panda\\student1.csv"
df = pd.read_csv(file)

#Creating class for pandas method and function
class dataframe:
    def __init__(self, filename, dataframe):
            self.filename = filename
            self.dataframe = dataframe
        
    def data_dimension(self):
        "This function return dimension of dataframe"
        try:
            print("\nDimension of our dataframe is : ", self.dataframe.ndim)
        except Exception as e:
            print(e)
       
    def data_type(self):
        "This function return type of dataframe"
        try:
            print("\nType of our dataframe is : " , type(self.dataframe))
        except Exception as e:
            print(e)
            
    
    def dataframe_size(self):
        "This function return size of dataframe"
        try:
            print("\nSize of our dataframe is : " , self.dataframe.size)
        except Exception as e:
            print(e)
    
    def first_five(self):
        "This function return first five entries of dataframe"
        try:
            print("\nFirst 5 entries from our dataframe is : \n" , self.dataframe.head())
        except Exception as e:
            print(e)
            
    def last_entries(self):
        "This function return last five entries of dataframe"
        try:
            print("\nLast 5 entries from our dataframe is : \n" , self.dataframe.tail())
        except Exception as e:
            print(e)

    def information(self):
        "This information return all information about dataframe"
        try:
            print("\nInformation about dataframe dataframe is : \n" ,self.dataframe.info())
        except Exception as e:
            print(e)
        
    def label_list(self):
        "This information return list of labels from dataframe"
        try:
            print("\nLabel of our dataframe is : \n", self.dataframe.axes)
        except Exception as e:
            print(e)
    
    def empty_object(self):
        "This information return list of labels from dataframe"
        try:
            #Returns the Boolean value saying whether the Object is empty or not. 
            #True indicates that the object is empty.
            print("\nGiven dataframe contain empty series or not : \n", self.dataframe.empty)
        except Exception as e:
            print(e)
    
    def shape_of_dataframe(self):
        "This function return shape of dataframe"
        try:
            print("\nShape of our dataframe is : " , self.dataframe.shape)
        except Exception as e:
            print(e)
    
    def random_dataframe(self):
        "This function return random sample"
        try:
            print("\nRandom entries from our dataframe is : \n" , self.dataframe.sample())
        except Exception as e:
            print(e)
            
    def tranpose_dataframe(self):
        "This function return transpose of dataframe"
        try:
            print("\nTranspose of dataframe is : \n" , self.dataframe.T)
        except Exception as e:
            print(e)
        
#object of the class 
d = dataframe(file, df)

#Calling functions
d.data_dimension()
d.data_type()
d.dataframe_size()
d.first_five()
d.last_entries()
d.information()
d.label_list()
d.empty_object()
d.shape_of_dataframe()
d.random_dataframe()
d.tranpose_dataframe()
```

# class  for Pandas cleaning data 

It consists of 
1. Empty cells
2. remove empty rows
3. Replace empty values
4. Replace only for specified column
5. Replace Using Mean, Median, or Mode
6. Data of Wrong Format
7. Fixing Wrong Data
8. Discovering Duplicates
9. Removing Duplicates

```python
#importing module
import pandas as pd

#File path for functionalits of panda
file = "F:\\PGD_Data_Science\\Lecture_video\\panda\\dirtydata.csv"
df = pd.read_csv(file)
x = int(input("Enter value to fill empty entries : "))

#Creating class for pandas method and function
class cleaning_data:
    def __init__(self, filename, dataframe, value):
            self.filename = filename
            self.dataframe = dataframe
            self.value = value
            
    def check_empty(self):
        "This function check for empty entries"
        try:
            print("\nDataFrame is : \n" , self.dataframe.isnull())
        except Exception as e:
            print(e)
            
    def count_empty(self):
        "This function count entry of empty entries"
        try:
            print("\nTotal empty value is : \n" , self.dataframe.isnull().sum())
        except Exception as e:
            print(e)
            
    def remove_empty(self):
        "This function count entry of empty entries"
        try:
            new_df = self.dataframe.dropna()
            print("\nDataframe after removal of empty entries : \n" , new_df)
        except Exception as e:
            print(e)
        
    def fill_empty(self):
        "This function fill of empty entries"
        try:
            new_df = self.dataframe.fillna(self.value)
            print("\nDataframe after removal of empty entries : \n" , new_df)
        except Exception as e:
            print(e)
        

cd = cleaning_data(file, df, x)

cd.check_empty()
cd.count_empty()
cd.remove_empty()
cd.fill_empty()
```
